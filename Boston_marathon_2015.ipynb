## Import packages

```python
import pandas as pd
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

## Read in the data

```python
Boston = pd.read_csv(pwd + "/marathon_results_2015.csv")
Boston
```

```python
# list all columns in the dataset
list(Boston.columns)
```

```python
# see basic descriptive statistics of numerical columns
Boston.describe()
```

```python
# find the data types of all the columns
Boston.dtypes
```

## Clean the data

```python
# rename the first column to "Index"
Boston.rename(columns={"Unnamed: 0": "Index"}, inplace=True)
Boston
```

```python
# Get rid of unnecessary columns: Index, Unnamed: 9, Proj Time; these columns were either empty or did not provide any value to the dataset
Boston.drop(columns=["Index", "Unnamed: 9", "Proj Time"], inplace=True)
Boston
```

## Explore/analyze the data

```python
# find the top 10 finishers from the U.S.
Boston[Boston["Country"] == "USA"].nsmallest(n=10, columns=["Overall"])
```

```python
# find the countries with the highest number of participants
Boston.groupby("Country")["Bib"].count().reset_index().sort_values(by="Bib", ascending=False)
```

```python
# find the top 10 states in the U.S. by number of participants
Boston[Boston["Country"] == "USA"].groupby("State")["Bib"].count().reset_index().sort_values(by="Bib", ascending=False).nlargest(n=10, columns=["Bib"])
```

```python
# look at just the half marathon times with name
Boston.loc[:, ["Name", "Half"]]
```

```python
# Create age groups
Boston["Age_group"] = np.select(
    [
        Boston["Age"].between(18, 34, inclusive=True),
        Boston["Age"].between(35, 39, inclusive=True),
         Boston["Age"].between(40, 44, inclusive=True),
          Boston["Age"].between(45, 49, inclusive=True),
           Boston["Age"].between(50, 54, inclusive=True),
            Boston["Age"].between(55, 59, inclusive=True),
             Boston["Age"].between(60, 64, inclusive=True),
              Boston["Age"].between(65, 69, inclusive=True),
               Boston["Age"].between(70, 74, inclusive=True),
                Boston["Age"].between(75, 79, inclusive=True),
                 Boston["Age"].between(80, 100, inclusive=True)
    ],
    [
        "18-34",
        "35-39",
        "40-44",
        "45-49",
        "50-54",
        "55-59",
        "60-64",
        "65-69",
        "70-74",
        "75-79",
        "80+"
    ]
)
Boston
```

```python
# change the max number of rows that we can view to 50, and the max number of columns we can view to 25
# find the top 50 male finishers in the 18 - 34 age group
pd.set_option("display.max_rows", 50)
pd.set_option('display.max_columns', 25)
Boston[(Boston["Age_group"] == "18-34") & (Boston["M/F"] == "M")].nsmallest(n=50, columns=["Overall"])
```

```python
# find the top 50 female finishers in the 18 - 34 age group
Boston[(Boston["Age_group"] == "18-34") & (Boston["M/F"] == "F")].nsmallest(n=50, columns=["Overall"])
```

```python
# find the top 25 finishers from Philadelphia
Boston[Boston["City"] == "Philadelphia"].nsmallest(n=25, columns=["Overall"])
```

```python
# find the top 25 finishers from Ohio
Boston[Boston["State"] == "OH"].nsmallest(n=25, columns=["Overall"])
```

```python
# find finishers from Tipp City
Boston[Boston["City"] == "Tipp City"]
```

```python
# find the number of male, female and total participants by age group, along with percentage of male and female participants
age_pivot = Boston.pivot_table(index="Age_group", columns="M/F", values= "Bib", aggfunc='count').reset_index()
age_pivot["Total"] = age_pivot["M"] + age_pivot["F"]
age_pivot["Percentage_Female"] = (age_pivot["F"] / age_pivot["Total"]) * 100
age_pivot["Percentage_Male"] = (age_pivot["M"] / age_pivot["Total"]) * 100
age_pivot
```

```python
# find the average age for each age group, by male and female participants
average_age = Boston.pivot_table(index= "Age_group", columns="M/F", values="Age", aggfunc='mean')
average_age
```
